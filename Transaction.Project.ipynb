## <STRONG> CLEANING OUR DATA </STRONG>

### <strong> check our data </strong>

```python
df
```

### <strong>Remove the duplicates</strong>

```python
df = df.drop_duplicates()
df
```

### <strong> Fix our columns (first and last names) using strip </strong>

```python
df["first_name"] = df["first_name"].str.strip("?1./")
df["last_name"] = df["last_name"].str.strip("_./1")
df
```

### <strong> Fix the transaction_type and category columns using "replace" </strong>

```python
df["transaction_type"] = df["transaction_type"].str.replace("Income","income")
df
```

```python
df["category"] = df["category"].str.replace("BONUS","bonus")
df
```

### <strong> Fill in the blancks/Null values </strong>

```python
df.loc[df["category"] == "rent", "transaction_type"] = "expence"
df.loc[df["category"] == "utilities", "transaction_type"] = "expence"
df.loc[df["category"] == "bonus", "transaction_type"] = "income"
df.loc[df["transaction_type"] == "income", "category"] = "bonus"
df
```

### <strong> Drop down the rows without any currency </strong>

```python
for x in df.index: 
   if df.loc[x, "currency"] == '':
       df.drop(x, inplace = True)
df
```

### <strong>Remove AGAIN the duplicates</strong>

```python
df = df.drop_duplicates()
```

### <strong> Dataset afte cleaning </strong>

```python
df
```

### <strong> Spelling correction </strong>

```python
df["transaction_type"] = df["transaction_type"].str.replace("expence","expense")
df
```

### <strong> Create a csv file </strong>

```python
df.to_csv("transactions.csv", index=False, encoding="utf-8")
```

## <strong> REGRESSION </strong>

```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
df
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
X = df[['Amount', 'transaction_type', 'currency']]
X_encoded = pd.get_dummies(X, columns=['transaction_type', 'currency'])
y = df['category']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
accuracy = model.score(X_test, y_test)
accuracy
```

```python
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)
```
